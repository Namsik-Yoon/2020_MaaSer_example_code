```python
#marker를 지도에 나타냄
#circle marker 지도에 나타냄
def add_marker(geo_list, geo_name, map_name):
    folium.Marker(geo_list, popup = geo_name).add_to(map_name)
    
def add_circle_marker(geo_list, geo_name, map_name, radius = 100
                      , color = '#3186cc', fill_color = "#3186cc"):
    folium.CircleMarker(geo_list,radius = radius, color = color,
                        fill_color= color, popop = geo_name).add_to(map_name)
```

```python
def direction_maker():
    """
출발점
도착점
mode = driving walking transit
waypoints 지정가능
return 값은 길찾기를 하고 난 raw한 값들

Travelling Salesman Problem 알고리즘으로 최적화된 길을 찾을 수 있다는 걸 봄
"""
    now = datetime.now()
    directions_result = gmaps.directions("US STEEL Edgar Thomson Works",
                                     "Rankin Christian Center",
                                     mode="driving",
                                     departure_time=now)
    return directions_result

directions_result = direction_maker()
directions_result
```
